```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("/content/drive/MyDrive/TECH A INTERN DATA ANALYSIS/task 1/olist_products_dataset.csv")
```

```python
# Check the dimensions of the dataset
df.shape
```

```python
df.head()
```

```python
df.describe()
```

```python
df.groupby('product_category_name').agg(
    {'product_weight_g': 'mean', 'product_length_cm': 'mean', 'product_height_cm': 'mean', 'product_width_cm': 'mean'})
```

```python
from matplotlib import pyplot as plt
df['product_weight_g'].plot(kind='hist', bins=20, title='product_weight_g')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
df.groupby('product_category_name')['product_weight_g'].mean().plot(kind='bar')
```

```python

df['product_name_lenght'].plot(kind='hist', bins=20, title='product_name_lenght')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python

correlation_matrix = df.corr()
print(correlation_matrix)

# Visualize the correlation matrix using a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.show()
```
